3. Preparación de los datos: 

```python
#- Los nombres de las columnas empiezan con mayúsculas, pon todos los nombres de las columnas en minúsculas. 
df = df.rename(columns=str.lower)
df
```

```python
#- Haciendo el análisis exploratorio te deberías haber dado cuenta de que algunas de las variables (`danceability`, `energy`, `key`, `loudness`, `speechiness`, `acousticness`, `instrumentalness`) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Crea una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto. 
col1 = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
```

```python
def to_float(valor):
    try:
        return float(valor.replace(",", "."))
    except:
        return np.nan
```

```python
for columna in col1:
    df[columna] = df[columna].apply(to_float)
```

```python
#- Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: ¿Son aquellos artistas que tienen más visitas los que más likes tienen? Para solucionar este ejercicio deberéis: 
#- Agrupa por artista y calcular la media de visitas y de likes.
consulta1 = df.groupby('artist').agg({'views' : ['mean'], 'likes' : ['mean']})
consulta1
```

```python
#- Quedate con los 10 artistas que más visitas han tenido.
consulta2 = consulta1.sort_values(by=('views', 'mean'), ascending=False)
top_10_views = consulta2.head(10)
top_10_views
```

```python
#- Quedate con los 10 artistas que más likes han tenido. 
consulta3 = consulta1.sort_values(by=('likes', 'mean'), ascending=False)
top_10_likes = consulta3.head(10)
top_10_likes
```

```python
# quiero saber si los artistas se repiten:
```

```python
artistas_views = set(top_10_views.index.tolist()) #hago un set de artistas con cada resultado
artistas_likes = set(top_10_likes.index.tolist())
```

```python
#busco intersecciones en los dos set:
artistas_comun = artistas_views.intersection(artistas_likes)
print(f'Coinciden {len(artistas_comun)} en ambas listas, en concreto: {artistas_comun}')
#podemos decir que hay una relación
```

```python
#- **BONUS** Haced lo mismo para los datos de tipo "single". 
consulta1_single = df_single.groupby('artist').agg({'views' : ['mean'], 'likes' : ['mean']}) #uso el dataframe de single crado en un ejercicio anterior y repito todo el código
consulta1_single.head(5)
```

```python
consulta2_single = consulta1_single.sort_values(by=('views', 'mean'), ascending=False)
top_10_views_single = consulta2_single.head(10)
top_10_views_single
```

```python
consulta3_single = consulta1_single.sort_values(by=('likes', 'mean'), ascending=False)
top_10_likes_single = consulta3_single.head(10)
top_10_likes_single
```

```python
artistas_views_single = set(top_10_views_single.index.tolist()) #hago un set de artistas con cada resultado
artistas_likes_single = set(top_10_likes_single.index.tolist())
artistas_comun_single = artistas_views_single.intersection(artistas_likes_single)
print(f'Coinciden {len(artistas_comun_single)} en ambas listas, en concreto: {artistas_comun_single}')
#coinciden incluso más artistas si filtramos por single
```

```python
#- Ahora queremos analizar la colaboración entre artistas en la industria musical, se tienen datos de las colaboraciones realizadas en los últimos años. Se desea conocer cuáles son los artistas que han colaborado en mayor cantidad de canciones y generar un *ranking* con los diez primeros. ¿Qué artistas con los que más colaboraciones tienen? Para esto tendréis que:
#- Filtra los datos para quedarte solo con aquellos datos donde en la columna "track" tengamos "feat".
df_feat = df[df['track'].str.contains('feat', case=False)]
df_feat.head(3)
```

```python
#- Cuenta cuantas canciones tiene cada artista utilizando el *dataframe* generado en el paso anterior.
df_feat_suma = df_feat.groupby('artist')['track'].count().sort_values(ascending=False).reset_index()
df_feat_ranking = pd.DataFrame(df_feat_suma.head(10))
df_feat_ranking
```

- Debes contestar a las siguientes preguntas:

```python
#- ¿Cuáles son los diez artistas que más han colaborado  según los datos que tenemos?
print(f'Los diez artistas que más han colaborado son: {df_feat_ranking["artist"].tolist()}')
```

```python
#- ¿Cuál es el número de canciones en los que ha colaborado el artista que ocupa el primer lugar en el *ranking*?
print(f'El número de canciones en los que ha colaborado el artista que ocupa el primer lugar en el *ranking*: {df_feat_ranking.iloc[0,1]}')

#10 canciones
```

```python
#- ¿Cuál es el artista que ocupa el décimo lugar en el *ranking* y en cuántas canciones ha colaborado?
print(f'El artista que ocupa el décimo lugar en el *ranking* es {df_feat_ranking.iloc[-1,0]} y ha colaborado en {df_feat_ranking.iloc[-1,1]} canciones')
```
